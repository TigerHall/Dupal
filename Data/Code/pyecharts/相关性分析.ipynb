相关性热力图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")  # 过滤掉警告的意思
```

```python
# 数据处理

df = pd.read_csv('test.csv')
df.head()
# print(df)
new_df = df.corr()
print(new_df)#打印值
```

```python
import holoviews as hv
from holoviews import opts

data = [(i, chr(97+j),  i*j) for i in range(5) for j in range(5) if i != j]

hm = hv.HeatMap(data).sort()
# hm.opts(xticks=None,)
```

```python
# 出图设置

# 图片显示中文
plt.rcParams['font.sans-serif'] = ['SimHei']
# 正常显示符号
plt.rcParams['axes.unicode_minus'] = False

# 设置横纵坐标的名称及热力图名称以及对应字体格式
font1 = {
    'family': 'SimHei',
    'weight': 'normal',
    'size': 100,
}

# 调整画布大小
ax = plt.subplots(figsize=(60, 60), dpi=200)

# 画热力图   annot=True 表示显示系数
ax = sns.heatmap(
    new_df,
    square=True,
    annot=True,
    # 网格内数字大小
    annot_kws={"fontsize": 60}
)

# colorbar
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=60)

# 设置坐标字体大小
plt.xticks(fontsize=60)
plt.yticks(fontsize=60)

# 热力图名称
ax.set_title('洋中脊同位素相关性图', font1)

# 出图
plt.savefig('洋中脊同位素相关性.png')
```
